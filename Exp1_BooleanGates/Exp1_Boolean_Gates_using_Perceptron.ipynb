```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# OR gate data
or_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [1, 1, 1, 0]
})

# AND gate data
and_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [1, 0, 0, 0]
})

# NAND gate data
nand_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [0, 1, 1, 1]
})

# NOR gate data
nor_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [0, 0, 0, 1]
})

# XOR gate data
xor_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [0, 1, 1, 0]
})

# XNOR gate data
xnor_data = pd.DataFrame({
    'input1': [1, 1, 0, 0],
    'input2': [1, 0, 1, 0],
    'output': [1, 0, 0, 1]
})
```

```python
sns.scatterplot(x=and_data['input1'], y=and_data['input2'], hue=and_data['output'], s=200)
plt.title('AND Gate')
plt.show()

sns.scatterplot(x=or_data['input1'], y=or_data['input2'], hue=or_data['output'], s=200)
plt.title('OR Gate')
plt.show()

sns.scatterplot(x=nor_data['input1'], y=nor_data['input2'], hue=nor_data['output'], s=200)
plt.title('NOR Gate')
plt.show()

sns.scatterplot(x=nand_data['input1'], y=nand_data['input2'], hue=nand_data['output'], s=200)
plt.title('NAND Gate')
plt.show()

sns.scatterplot(x=xor_data['input1'], y=xor_data['input2'], hue=xor_data['output'], s=200)
plt.title('XOR Gate')
plt.show()
```

```python
# OR gate
or_weights = [1, 1]
or_bias = -0.5

# AND gate
and_weights = [1, 1]
and_bias = -1.5

# NAND gate
nand_weights = [-1, -1]
nand_bias = 1.5

# NOR gate
nor_weights = [-1, -1]
nor_bias = 0.5

# XOR gate (requires two layers)
xor_weights_1 = [[1, 1], [-1,-1]] #weights for the first layer
xor_bias_1 = [-0.5, 0.5] # bias for the first layer
xor_weights_2 = [1, -1] #weights for second layer
xor_bias_2 = -0.5 # bias for second layer


# XNOR gate (requires two layers)
xnor_weights_1 = [[1, 1], [-1,-1]] #weights for the first layer
xnor_bias_1 = [-0.5, 0.5] # bias for the first layer
xnor_weights_2 = [-1, 1] #weights for second layer
xnor_bias_2 = 0.5 # bias for second layer


print("OR Gate: weights =", or_weights, ", bias =", or_bias)
print("AND Gate: weights =", and_weights, ", bias =", and_bias)
print("NAND Gate: weights =", nand_weights, ", bias =", nand_bias)
print("NOR Gate: weights =", nor_weights, ", bias =", nor_bias)
print("XOR Gate: weights_layer1 =", xor_weights_1, ", bias_layer1 =", xor_bias_1, ", weights_layer2 =", xor_weights_2, ", bias_layer2 =", xor_bias_2)
print("XNOR Gate: weights_layer1 =", xnor_weights_1, ", bias_layer1 =", xnor_bias_1, ", weights_layer2 =", xnor_weights_2, ", bias_layer2 =", xnor_bias_2)
```

```python

def OR(or_data, or_weights, or_bias):
    fx = np.dot(or_data.iloc[:, :2], or_weights) + or_bias
    if fx >= 0:
        return 1
    else:
        return 0

print("OR Gate:")
for index, row in or_data.iterrows():
    inputs = row[['input1', 'input2']]
    expected_output = row['output']
    perceptron_output = OR(pd.DataFrame([inputs]), or_weights, or_bias)
    print(f"Inputs: {inputs.to_list()}, Predicted Output: {perceptron_output}, Expected Output: {expected_output}")



def AND(and_data, and_weights, and_bias):
    fx = np.dot(and_data.iloc[:, :2], and_weights) + and_bias
    if fx >= 0:
        return 1
    else:
        return 0

print("\nAND Gate:")
for index, row in and_data.iterrows():
    inputs = row[['input1', 'input2']]
    expected_output = row['output']
    perceptron_output = AND(pd.DataFrame([inputs]), and_weights, and_bias)
    print(f"Inputs: {inputs.to_list()}, Predicted Output: {perceptron_output}, Expected Output: {expected_output}")


# NAND gate
nand_weights = [-1, -1]
nand_bias = 1.5

def NAND(nand_data, nand_weights, nand_bias):
    fx = np.dot(nand_data.iloc[:, :2], nand_weights) + nand_bias
    if fx >= 0:
        return 1
    else:
        return 0

print("\nNAND Gate:")
for index, row in nand_data.iterrows():
    inputs = row[['input1', 'input2']]
    expected_output = row['output']
    perceptron_output = NAND(pd.DataFrame([inputs]), nand_weights, nand_bias)
    print(f"Inputs: {inputs.to_list()}, Predicted Output: {perceptron_output}, Expected Output: {expected_output}")

# NOR gate
nor_weights = [-1, -1]
nor_bias = 0.5

def NOR(nor_data, nor_weights, nor_bias):
    fx = np.dot(nor_data.iloc[:, :2], nor_weights) + nor_bias
    if fx >= 0:
        return 1
    else:
        return 0

print("\nNOR Gate:")
for index, row in nor_data.iterrows():
    inputs = row[['input1', 'input2']]
    expected_output = row['output']
    perceptron_output = NOR(pd.DataFrame([inputs]), nor_weights, nor_bias)
    print(f"Inputs: {inputs.to_list()}, Predicted Output: {perceptron_output}, Expected Output: {expected_output}")

```

```python

def plot_decision_boundary(data, weights, bias, title):
    x_values = np.linspace(min(data['input1']) - 0.5, max(data['input1']) + 0.5, 100)
    y_values = - (weights[0] * x_values + bias) / weights[1]
    sns.scatterplot(x=data['input1'], y=data['input2'], hue=data['output'], s=200)
    plt.plot(x_values, y_values, 'r--')  # Plot the decision boundary
    plt.title(title)
    plt.xlim([-0.5, 1.5])  # Set x-axis limits for better visualization
    plt.ylim([-0.5, 1.5])  # Set y-axis limits for better visualization
    plt.show()


plot_decision_boundary(and_data, and_weights, and_bias, "AND Gate")
plot_decision_boundary(or_data, or_weights, or_bias, "OR Gate")
plot_decision_boundary(nand_data, nand_weights, nand_bias, "NAND Gate")
plot_decision_boundary(nor_data, nor_weights, nor_bias, "NOR Gate")

```
